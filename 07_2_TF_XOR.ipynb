### VERY Important  : XOR

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

tf.set_random_seed(777)  # for reproducibility
```

* Hyperparameters

```python
learning_rate = 0.1 #촘촘히,느리게 학습
nb_epoch = 10000 #학습을 총 만 번 하기 
```

* Dataset

```python
x_data = [[0, 0],
          [0, 1],
          [1, 0],
          [1, 1]]

y_data = [[0],
          [1],
          [1],
          [0]]
```

```python
W = tf.placeholder(tf.float32)
```

```python
x_data = np.array(x_data, dtype=np.float32)
y_data = np.array(y_data, dtype=np.float32)
```

```python
X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 1])

W1 = tf.Variable(tf.random_normal([2, 2]), name='weight1')#입력이 2개 연결하는선이 2개 총 선 4개, random_normal
b1 = tf.Variable(tf.random_normal([2]), name='bias1')#입력 2개를 받는 것이 2개 동그라미4개 
layer1 = tf.sigmoid(tf.matmul(X, W1) + b1)

W2 = tf.Variable(tf.random_normal([2, 1]), name='weight2')
b2 = tf.Variable(tf.random_normal([1]), name='bias2')#입력 2개를 받는 것이 1개 동그라미 총 5개 
hypothesis = tf.sigmoid(tf.matmul(layer1, W2) + b2)#sigmoid 
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
train = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)
```

```python
# Launch graph
sess = tf.Session()
```

```python
# TensorFlow 변수들(variables) 초기화 (Initialization)
sess.run(tf.global_variables_initializer())
```

```python
for i in range(nb_epoch+1):
    sess.run(train, feed_dict={X: x_data, Y: y_data})#에포크가 돌 때 데이터가 돌려진다.
    
    c1 = sess.run(cost, feed_dict={X: x_data, Y: y_data})
    plt.plot(i,c1,'b.')
```

```python
Step_X=[]
Cost_Y=[]

for i in range(nb_epoch+1):
    sess.run(train, feed_dict={X: x_data, Y: y_data})#에포크가 돌 때 데이터가 돌려진다.

    if i % 1000 ==0:
        c1 = sess.run(cost, feed_dict={X: x_data, Y: y_data})
        print('step={} / cost={}'.format(i, c1))
        Cost_Y.append(c1)
        Step_X.append(i)
  
plt.plot(Step_X,Cost_Y)
plt.grid
plt.title('Cost')
```
