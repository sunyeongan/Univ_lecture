```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```


**X and Y data**

```
x_train = [1, 2, 3, 4 ,5]
#y_train = [2, 4, 6, 8, 10]
y_train = [2+0.1+3, 4-0.3+3, 6+0.15+3, 8+0.1+3, 10-0.12+3] #  Add some noise 
```


**Initialization**

```
#W = tf.Variable(tf.random_normal([1]), name='weight')
#b = tf.Variable(tf.random_normal([1]), name='bias')
w0 = 12;
b0 = 50;

W = tf.Variable(w0*tf.ones([1]), name='weight')
b = tf.Variable(b0*tf.ones([1]), name='bias')
```

**Our hypothesis XW+b**

```
hypothesis = x_train * W + b
```

**cost/loss function**

```
cost = tf.reduce_mean(tf.square(hypothesis - y_train)) #스퀘어 = 제곱//평균계산해서 작게 만드는 값을 학습하는 함수?
```

**Optimizer**

```
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

**Launch the graph in a session**

```
sess = tf.Session()
```

**Initializes global variables in the graph.** 

```
sess.run(tf.global_variables_initializer())#변수 안에 텐서연산 실행 
```

```
vw = []
vb = []
for step in range(2001):
    sess.run(train)
    w1 = sess.run(W)[0] # slope
    b1 = sess.run(b)[0] # bias
    vw.append(w1)
    vb.append(b1)
    if step % 100 == 0:
        print(step, sess.run(cost), w1, b1)
```

**Complete training**

```
plt.plot(vw)
```

```
plt.plot(vb)
```

```
w1 = sess.run(W)[0] # slope
b1 = sess.run(b)[0] # bias
str1 = 'y = ' + str(w1) +'x + ' + str(b1)
print(w1, b1)
print(str1)
```

```

plt.figure(1)
plt.plot(x_train, y_train,'o')

x1 = np.linspace(np.min(x_train)-1, np.max(x_train)+1)
#linspace ->선형구간 혹은 로그 구간을 지정한 만큼 구간의 수만큼 분할한다.
y1 = w1*x1 + b1
plt.plot(x1, y1)
plt.grid()
plt.title(str1)
```
