```
import tensorflow as tf
mnist = tf.keras.datasets.mnist
```

```
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0 ,x_test / 255.0
```

```
x_train.shape
```

```
x_test.shape
```

```
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),#한층으로 펴다 ->784
  tf.keras.layers.Dense(256, activation=tf.nn.relu),#거미줄 
  tf.keras.layers.Dense(256, activation=tf.nn.relu),#거미줄  
  
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)# 밀도,mnist lable 의 확률 (담는그릇) 
])
model.compile(optimizer='adam',#문제풀기 정의 adam은 절반이 모멘텀(공굴리기)
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
F=784
D_3=512
D_4=10
sum1 = F*D_3+D_3#F*D_3은 weight고 +D_3는 bias를 덧셈한다.
print(sum1)
sum2 = D_3*D_4+D_4
print(sum2)
print(sum1+sum2)
```

```
model.summary() # #은 숫자,개수 시험시험시험시험 dense -> 거미줄 
```

```

model.fit(x_train, y_train, epochs=5)
```

### *Real World Challenge*: Large difference between training and testing set accuracy

#### Test accuracy

```
model.evaluate(x_test, y_test)#test set으로 98프로가 나왔다 
```

#### Trainning accuracy

```
model.evaluate(x_train, y_train)
```
