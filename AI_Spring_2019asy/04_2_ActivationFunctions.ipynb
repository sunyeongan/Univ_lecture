```
import numpy as np
import matplotlib.pylab as plt
```

```
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```
def relu(x):
    return np.maximum(0, x)
```

```
def step_function(x):
    return np.array(x > 0, dtype=np.int)
```

```
X = np.arange(-5.0, 5.0, 0.1)
```

```
Y1 = relu(X)
Y2 = sigmoid(X)
Y3 = step_function(X)
```

```
plt.plot(X, Y1)
plt.ylim(-0.1, 1.1)
plt.title('relu')
```

```
plt.plot(X, Y2)
plt.ylim(-0.1, 1.1)
plt.title('sigmoid')
```

```
plt.plot(X, Y3)
plt.ylim(-0.1, 1.1)
plt.title('step function')
```
